```python
import pandas as pd
import numpy as np 
import re # regular expression
from nltk.corpus import stopwords # natural language tool kit and corpus for the body of the text (stopwords are like a,an ,the e.t.c)
from nltk.stem.porter import PorterStemmer #stem takes a word and remove prefix and suffix of the word and return the root ot the word
from sklearn.feature_extraction.text import TfidfVectorizer # to convert text into feature vector
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

```

```python
import nltk 
nltk.download('stopwords')
```

```python
data = pd.read_csv("WELFake_Dataset.csv")
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data=data.fillna('')
```

```python
port_stem=PorterStemmer()
```

```python
def stemming(content):
    stemmed_content=re.sub('[˄a-zA-Z]',' ',content)
    stemmed_content = content.lower()
    stemmed_content = stemmed_content.split()
    stemmed_content = [
        port_stem.stem(word) 
        for word in stemmed_content 
        if word not in stopwords.words('english')
    ]
    stemmed_content = ' '.join(stemmed_content)
    return stemmed_content
```

```python
data['text']=data['text'].apply(stemming)
```

```python
print(data['text'])
```

```python
x=data['text'].values
y=data['label'].values
```

```python
#convert textual data into numerical data meanigful using vectorizer
vectorizer =TfidfVectorizer() # Tf->Term frequenze idf-> inverse document frequenze it basically count the no of times a particular word repating
vectorizer.fit(x)

x=vectorizer.transform(x)
```

```python
print(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
predict1=model.predict(x_test)
```

```python
accuracy=accuracy_score(predict1,y_test)
```

```python
print(accuracy)
```

```python
X_new=x_test[1]
prediction=model.predict(X_new)
print(prediction)

if(prediction[0]==0):
    print("The news is real")
else:
    print("The news is Fake")
```
